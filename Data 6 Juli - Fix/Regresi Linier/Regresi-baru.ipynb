```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Plot PSD Regresi Linear

```python
X = np.array([0.00284585, 0.00118143, 0.00000000, 0.00000035]).reshape(-1,1)
Y = np.array([30, 50, 70, 90])
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
koefisien_kemiringan = model.coef_[0]
konstanta = model.intercept_
```

```python
print("Koefisien Kemiringan: ", koefisien_kemiringan)
print("Konstanta: ", konstanta)
```

```python
Y_prediksi = model.predict(X)
```

```python
plt.scatter(X, Y, label='Data')
plt.plot(X, Y_prediksi, color='red', label='Regresi Linear')
plt.xlabel('Nilai Puncak Rata-rata')
plt.ylabel('Densitas Model Tulang')
plt.legend()
plt.grid(True)
plt.title("REGRESI LINEAR RATA-RATA PUNCAK PSD SISI ATAS")
plt.show()
```

# Plot FFT Regresi Polinomial

```python
# Menghitung nilai R^2
mean_y = np.mean(Y)
predicted_y = regression_line(X)
r_squared = 1 - (np.sum((Y - predicted_y)**2) / np.sum((Y - mean_y)**2))

print("Nilai R^2:", r_squared)
```
