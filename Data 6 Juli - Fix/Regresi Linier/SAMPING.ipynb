```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([0.032401, 0.008570, 0.004040, 0.007041]).reshape(-1, 1)
Y = np.array([30, 50, 70, 90])
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
koefisien_kemiringan = model.coef_[0]
konstanta = model.intercept_
```

```python
print("Koefisien Kemiringan: ", koefisien_kemiringan)
print("Konstanta: ", konstanta)
```

```python
Y_prediksi = model.predict(X)
```

```python
plt.scatter(X, Y, label='Data')
plt.plot(X, Y_prediksi, color='red', label='Regresi Linear')
plt.xlabel('Nilai Puncak pada Sisi-sisi Tulang')
plt.ylabel('Densitas Tulang')
plt.legend()
plt.grid(True)
plt.show()
```
