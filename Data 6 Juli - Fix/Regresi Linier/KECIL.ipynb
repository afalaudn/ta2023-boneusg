```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([0.044861, 0.010857, 0.0231494, 0.005385]).reshape(-1, 1)
Y = np.array([30, 50, 70, 90])
```

```python
model = LinearRegression()
```

```python
model.fit(X, Y)
```

```python
koefisien_kemiringan = model.coef_[0]
konstanta = model.intercept_
```

```python
print("Koefisien Kemiringan: ", koefisien_kemiringan)
print("Konstanta: ", konstanta)
```

```python
Y_prediksi = model.predict(X)
```

```python
plt.scatter(X, Y, label='Data')
plt.plot(X, Y_prediksi, color='red', label='Regresi Linear')
plt.xlabel('Nilai Puncak Rata-rata')
plt.ylabel('Densitas Tulang')
plt.legend()
plt.grid(True)
plt.title("REGRESI LINEAR RATA-RATA PUNCAK PSD SISI KECIL")
plt.show()
```
